```python
import cv2
```

```python
import tensorflow as tf
from keras.models import load_model
import numpy as np
from pygame import mixer
```

```python
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eye_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
model=load_model(r'D:\Model\DD_Project\model checkpoint\model.keras')
```

```python
import tkinter as tk
from tkinter import messagebox

# Create a Tkinter window
root = tk.Tk()
root.withdraw()  # Hide the root window

# Show a pop-up message with instructions
messagebox.showinfo("Instructions", "To exit the application, press 'q'.")

import time
mixer.init()
sound=mixer.Sound(r'alarm.wav')
cap=cv2.VideoCapture(0)
# Define update interval (in seconds)
update_interval = 1  # Update score and status every 1 second

# Initialize variables for score smoothing
score = 0
last_update_time = time.time()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(frame, (ex, ey), (ex + ew, ey + eh), (255, 0, 0), 3)
        eye = frame[ey:ey + eh, ex:ex + w]
        eye = cv2.resize(eye, (80, 80))
        eye = eye / 255
        eye = np.expand_dims(eye, axis=0)
        
        # Preprocess the eye image and make predictions
        prediction = model.predict(eye)
        
    # Update score and status every specified interval
    current_time = time.time()
    if current_time - last_update_time >= update_interval:
        last_update_time = current_time
        if prediction[0][0] > 0.30:
            score += 2
            if score >=3:
                try:
                    sound.play()
                except:
                    pass
            status_text = 'Closed'
        elif prediction[0][1] > 0.50:
            if score > 0:
                score -= 1
            status_text = 'Open'
    
    # Display the score and frame
    cv2.putText(frame, 'Score: ' + str(score), (10, height - 20), cv2.FONT_HERSHEY_COMPLEX, 0.8, (255, 255, 255), 1)
    cv2.putText(frame, 'Eyes: ' + status_text, (10, height - 50), cv2.FONT_HERSHEY_COMPLEX, 0.8, (255, 255, 255), 1)
    cv2.imshow('Drowsiness Detection', frame)
    
    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
